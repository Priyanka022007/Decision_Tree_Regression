```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Dataset

ds = pd.read_csv("Position_Salaries.csv")
ds
```

```python
# Splitting data into inputdata and output data

x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
y
```

```python
# Fitting Decision Tree Regression model

from sklearn.tree import DecisionTreeRegressor,plot_tree
reg = DecisionTreeRegressor(random_state =1)
reg.fit(x,y)
```

```python
# Predict for myinput

myinput = pd.DataFrame(data=[[4]],columns =["Level"])
result=reg.predict(myinput)
print("Your result :",round(result[0]))
```

```python
# Visualization

plt.figure(figsize=(20,50))
plot_tree(reg,feature_names = ["Level"],filled =True)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
# For User input

level = float(input("Enter Level :"))
myinput = pd.DataFrame(data =[[level]],columns=["Level"])
result = reg.predict(myinput)
print(f"Predicted Salary for Level:{level} : {result[0]}")
```
